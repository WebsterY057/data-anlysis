```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
x = np.linspace(0, 10, 100)  # 0到10之间均匀分布的100个点
y = np.sin(x)  # 正弦函数

# 创建图形和坐标轴
plt.figure(figsize=(10, 6))  # 设置图形大小，单位英寸

# 绘制线图
plt.plot(x, y, 
         color='blue',        # 线条颜色
         linewidth=2,         # 线条宽度
         linestyle='-',       # 线条样式：'-'实线，'--'虚线，':'点线
         marker='o',          # 标记样式
         markersize=4,        # 标记大小
         label='sin(x)')      # 图例标签

# 添加标题和标签
plt.title('sin picture', fontsize=16, fontweight='bold')
plt.xlabel('Xaixs', fontsize=12)
plt.ylabel('Yaixs', fontsize=12)

# 添加网格
plt.grid(True, alpha=0.3)  # alpha控制透明度

# 添加图例
plt.legend()

# 设置坐标轴范围
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# 显示图形
plt.tight_layout()  # 自动调整子图参数
plt.show()
```

```python
# 创建随机数据
np.random.seed(42)  # 设置随机种子确保可重复性
n = 100
x = np.random.randn(n)  # 正态分布随机数
y = 2 * x + np.random.randn(n) * 0.5  # 添加噪声的线性关系

# 创建散点图
plt.figure(figsize=(10, 6))

plt.scatter(x, y,
            s=50,           # 点的大小
            c=x,            # 颜色映射，根据x值变化
            cmap='viridis', # 颜色映射方案
            alpha=0.6,      # 透明度
            edgecolors='black',  # 边缘颜色
            linewidth=0.5) # 边缘线宽

# 添加颜色条
plt.colorbar(label='Xvalues')

# 添加回归线
z = np.polyfit(x, y, 1)  # 一次多项式拟合
p = np.poly1d(z)
plt.plot(sorted(x), p(sorted(x)), "r--", label=f'回归线: y={z[0]:.2f}x+{z[1]:.2f}')

plt.title('scotter exp', fontsize=16)
plt.xlabel('X variable', fontsize=12)
plt.ylabel('Y variable', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 加载数据集
tips = sns.load_dataset('tips')

# 设置中文字体和负号显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 创建2x2的子图
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
fig.suptitle('分布分析图表', fontsize=16, fontweight='bold', y=1.02)

# 1. 直方图 (Histogram) - 左上
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('消费金额分布直方图', fontsize=12)
axes[0, 0].set_xlabel('消费金额')
axes[0, 0].set_ylabel('频数')
axes[0, 0].grid(True, alpha=0.3)

# 2. 核密度估计图 (KDE Plot) - 右上
sns.kdeplot(data=tips, x='total_bill', hue='time', fill=True, ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('不同时间段的消费分布', fontsize=12)
axes[0, 1].set_xlabel('消费金额')
axes[0, 1].set_ylabel('密度')
axes[0, 1].legend(title='用餐时间')
axes[0, 1].grid(True, alpha=0.3)

# 3. 箱线图 (Box Plot) - 左下
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker', ax=axes[1, 0], palette='husl')
axes[1, 0].set_title('每日消费箱线图（按是否吸烟分组）', fontsize=12)
axes[1, 0].set_xlabel('星期')
axes[1, 0].set_ylabel('消费金额')
axes[1, 0].legend(title='是否吸烟')
axes[1, 0].grid(True, alpha=0.3, axis='y')

# 4. 散点图 (Scatter Plot) - 右下
scatter = axes[1, 1].scatter(tips['total_bill'], tips['tip'], 
                             c=tips['size'],  # 颜色按用餐人数
                             s=tips['size']*20,  # 大小按用餐人数
                             alpha=0.6,
                             cmap='viridis')
axes[1, 1].set_title('消费金额与小费关系（大小颜色表示用餐人数）', fontsize=12)
axes[1, 1].set_xlabel('消费金额')
axes[1, 1].set_ylabel('小费')
axes[1, 1].grid(True, alpha=0.3)

# 添加颜色条
cbar = plt.colorbar(scatter, ax=axes[1, 1])
cbar.set_label('用餐人数')

plt.tight_layout()
plt.show()

# 5. 联合分布图 (Joint Plot) - 独立图形
print("\n" + "="*60)
print("联合分布图 (单独显示)")
print("="*60)

g = sns.jointplot(data=tips, x='total_bill', y='tip', 
                  kind='scatter',  # 可选: 'scatter', 'hist', 'hex', 'kde', 'reg', 'resid'
                  hue='smoker',    # 按吸烟者分组
                  height=6,        # 图形高度
                  alpha=0.6,       # 透明度
                  palette='Set2',
                  space=0.2)       # 主图与边缘图间距

# 添加回归线
sns.regplot(data=tips[tips['smoker'] == 'Yes'], 
            x='total_bill', y='tip', 
            scatter=False, ax=g.ax_joint, color='red', label='吸烟者回归线')
sns.regplot(data=tips[tips['smoker'] == 'No'], 
            x='total_bill', y='tip', 
            scatter=False, ax=g.ax_joint, color='blue', label='非吸烟者回归线')

g.ax_joint.legend()
g.fig.suptitle('消费金额与小费关系（按吸烟者分组）', y=1.02, fontsize=14)
g.set_axis_labels('消费金额', '小费')
plt.tight_layout()
plt.show()

# 6. 对子图 (Pair Plot) - 独立图形
print("\n" + "="*60)
print("变量对子图 (单独显示)")
print("="*60)

# 选择数值列
numeric_cols = ['total_bill', 'tip', 'size']
pair_data = tips[['total_bill', 'tip', 'size', 'sex', 'smoker']]

# 对子图
g = sns.pairplot(pair_data, 
                 hue='sex',        # 按性别分组
                 palette='husl',   # 颜色方案
                 diag_kind='kde',  # 对角线图形类型: 'hist' 或 'kde'
                 kind='scatter',   # 非对角线图形类型
                 plot_kws={'alpha': 0.6, 's': 50},  # 绘图参数
                 diag_kws={'fill': True},  # 对角线图参数
                 height=2.5)  # 每个子图的高度

g.fig.suptitle('数值变量对子图（按性别分组）', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 多组数据雷达图
fig = plt.figure(figsize=(12, 10))

# 创建多个子图
categories = ['数据分析', '编程能力', '业务理解', '沟通表达', '项目管理', '团队协作']
N = len(categories)

# 多个部门的数据
departments = ['技术部', '市场部', '产品部', '运营部']
data = {
    '技术部': [4.8, 4.9, 3.5, 3.8, 4.0, 4.2],
    '市场部': [3.2, 2.8, 4.7, 4.8, 4.1, 4.3],
    '产品部': [4.0, 3.8, 4.5, 4.2, 4.6, 4.1],
    '运营部': [3.8, 3.5, 4.2, 4.6, 4.3, 4.7]
}

# 创建2x2子图
for idx, (dept, values) in enumerate(data.items(), 1):
    ax = fig.add_subplot(2, 2, idx, polar=True)
    
    # 计算角度
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    values_ring = values + values[:1]
    angles += angles[:1]
    
    # 绘制雷达图
    ax.plot(angles, values_ring, 'o-', linewidth=2, color='#45B7D1')
    ax.fill(angles, values_ring, alpha=0.25, color='#45B7D1')
    
    # 设置标签
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=9)
    ax.set_ylim(0, 5)
    
    # 设置网格
    ax.grid(True, alpha=0.3)
    
    # 添加分数标签
    for i, (angle, value) in enumerate(zip(angles[:-1], values)):
        rotation_angle = np.degrees(angle)
        if rotation_angle > 90 and rotation_angle < 270:
            rotation_angle += 180
        ax.text(angle, value + 0.2, f'{value:.1f}', 
                ha='center', va='center', fontsize=9, fontweight='bold',
                rotation=rotation_angle, rotation_mode='anchor')
    
    ax.set_title(f'{dept}能力评估', fontsize=12, fontweight='bold', pad=20)

plt.suptitle('各部门能力评估雷达图', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle

# 创建日历数据
dates = pd.date_range('2023-01-01', '2023-12-31', freq='D')
values = np.random.randn(len(dates)).cumsum() + 100
values = values - values.min()  # 归一化到正值

# 创建DataFrame
df_calendar = pd.DataFrame({
    'date': dates,
    'value': values
})

# 提取年、月、日、星期
df_calendar['year'] = df_calendar['date'].dt.year
df_calendar['month'] = df_calendar['date'].dt.month
df_calendar['day'] = df_calendar['date'].dt.day
df_calendar['weekday'] = df_calendar['date'].dt.weekday
df_calendar['week'] = df_calendar['date'].dt.isocalendar().week

# 创建日历热力图
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
months = range(1, 13)
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

for ax, month in zip(axes.flatten(), months):
    # 筛选月份数据
    month_data = df_calendar[df_calendar['month'] == month]
    
    if not month_data.empty:
        # 创建矩阵
        pivot_table = month_data.pivot_table(
            index='weekday',
            columns='week',
            values='value',
            aggfunc='mean'
        )
        
        # 绘制热力图
        sns.heatmap(pivot_table, 
                   ax=ax,
                   cmap='YlOrRd',
                   cbar=False,
                   square=True,
                   linewidths=0.5,
                   linecolor='white')
    
    # 设置标题
    ax.set_title(month_names[month-1], fontweight='bold')
    ax.set_xlabel('')
    ax.set_ylabel('')
    
    # 设置刻度
    weekday_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    ax.set_yticks(np.arange(7) + 0.5)
    ax.set_yticklabels(weekday_names, rotation=0, fontsize=8)
    
    # 隐藏x轴刻度
    ax.set_xticks([])

plt.suptitle('2023年月度数据日历热力图', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
import squarify
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
data = {
    'category': ['Technology', 'Healthcare', 'Finance', 'Consumer', 'Energy', 'Industrial', 'Materials'],
    'value': [350, 280, 220, 180, 120, 90, 60],
    'growth': [12.5, 8.3, 5.7, 9.1, 3.4, 6.8, 2.9]
}

df = pd.DataFrame(data)

# 创建颜色映射
norm = plt.Normalize(df['growth'].min(), df['growth'].max())
colors = plt.cm.YlOrRd(norm(df['growth']))

# 创建图形和坐标轴
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# 创建树状图
squarify.plot(sizes=df['value'], 
              label=[f'{cat}\n${val}B\n{growth}%' 
                     for cat, val, growth in zip(df['category'], df['value'], df['growth'])],
              color=colors,
              alpha=0.8,
              text_kwargs={'fontsize': 10, 'fontweight': 'bold'},
              ax=ax)  # 指定axes参数

# 添加标题
ax.set_title('Market Value by Sector (with Growth Rate)', 
             fontsize=16, fontweight='bold', pad=20)

# 移除坐标轴
ax.axis('off')

# 添加颜色条
sm = plt.cm.ScalarMappable(cmap='YlOrRd', norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, orientation='horizontal', pad=0.05, aspect=50)
cbar.set_label('Growth Rate (%)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# 创建分层数据
data = {
    'sector': ['Tech']*4 + ['Finance']*3 + ['Healthcare']*3,
    'subsector': ['Software', 'Hardware', 'Internet', 'Semiconductor',
                  'Banking', 'Insurance', 'Investment',
                  'Pharma', 'Biotech', 'Medicare'],
    'company': ['Apple', 'Intel', 'Google', 'Nvidia',
                'JPMorgan', 'Berkshire', 'BlackRock',
                'Pfizer', 'Moderna', 'UnitedHealth'],
    'value': [2800, 240, 1800, 400,
              470, 700, 120,
              240, 180, 450]
}

df = pd.DataFrame(data)

# 创建旭日图
fig = px.sunburst(df,
                   path=['sector', 'subsector', 'company'],  # 分层路径
                   values='value',                           # 大小值
                   color='value',                            # 颜色值
                   color_continuous_scale='RdYlBu',          # 颜色方案
                   hover_data={'value': ':.0f'},             # 悬停数据格式
                   title='Market Value Sunburst Chart')

fig.update_layout(
    width=800,
    height=800,
    title_x=0.5,
    font=dict(size=12)
)

fig.show()
```

```python
import plotly.graph_objects as go

# 桑基图数据
labels = ["A", "B", "C", "D", "E", "F", "G", "H", "I"]

# 定义流
sources = [0, 0, 1, 1, 2, 2, 3, 4, 5, 6, 7]
targets = [3, 4, 3, 5, 4, 5, 6, 6, 7, 7, 8]
values =  [8, 4, 6, 2, 5, 3, 7, 5, 4, 6, 9]

# 定义颜色
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7',
          '#DDA0DD', '#98D8C8', '#F7DC6F', '#BB8FCE']

# 创建桑基图
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels,
        color=colors
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
        color=[colors[src] for src in sources]
    )
)])

fig.update_layout(
    title_text="销售渠道转化桑基图",
    font_size=12,
    width=1000,
    height=700
)

fig.show()
```

```python
# 创建示例数据
np.random.seed(42)
n_points = 50
categories = ['A', 'B', 'C', 'D']

data = pd.DataFrame({
    'x': np.random.randn(n_points) * 10 + 50,
    'y': np.random.randn(n_points) * 5 + 20,
    'size': np.random.uniform(100, 1000, n_points),
    'category': np.random.choice(categories, n_points),
    'value': np.random.uniform(1, 10, n_points)
})

# 创建颜色映射
category_colors = {'A': '#FF6B6B', 'B': '#4ECDC4', 'C': '#45B7D1', 'D': '#96CEB4'}

# 创建气泡图
plt.figure(figsize=(12, 8))

for category in categories:
    subset = data[data['category'] == category]
    plt.scatter(subset['x'], subset['y'],
                s=subset['size'],           # 气泡大小
                c=category_colors[category], # 颜色
                alpha=0.6,                  # 透明度
                edgecolors='black',         # 边缘颜色
                linewidths=0.5,             # 边缘线宽
                label=category)             # 图例标签

# 添加回归线
z = np.polyfit(data['x'], data['y'], 1)
p = np.poly1d(z)
plt.plot(data['x'].sort_values(), p(data['x'].sort_values()), 
         "k--", alpha=0.5, label='趋势线')

# 添加标签
for i, row in data.nlargest(5, 'size').iterrows():
    plt.annotate(f"{row['category']}\n({row['value']:.1f})", 
                 (row['x'], row['y']),
                 xytext=(5, 5),
                 textcoords='offset points',
                 fontsize=9,
                 fontweight='bold',
                 bbox=dict(boxstyle="round,pad=0.3", 
                          facecolor="white", 
                          alpha=0.7,
                          edgecolor='gray'))

plt.xlabel('X轴 (单位: 百万)', fontsize=12)
plt.ylabel('Y轴 (单位: %)', fontsize=12)
plt.title('气泡图示例: 大小表示规模，颜色表示类别', fontsize=16, fontweight='bold')

# 添加图例
legend = plt.legend(title='类别', loc='upper right')
legend.get_title().set_fontweight('bold')

# 添加网格
plt.grid(True, alpha=0.3, linestyle='--')

# 添加颜色条的替代（气泡大小图例）
sizes = [200, 500, 800]
for size in sizes:
    plt.scatter([], [], s=size, c='gray', alpha=0.6, 
                edgecolors='black', label=f'{size}')

# 创建第二个图例
handles, labels = plt.gca().get_legend_handles_labels()
legend1 = plt.legend(handles[:5], labels[:5], title='类别', 
                     loc='upper left', bbox_to_anchor=(1.02, 1))
plt.gca().add_artist(legend1)
plt.legend(handles[5:], labels[5:], title='气泡大小', 
           loc='upper left', bbox_to_anchor=(1.02, 0.7))

plt.tight_layout()
plt.show()
```

```python
# 创建极坐标图
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='polar')

# 数据
categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
N = len(categories)

# 创建多组数据
theta = np.linspace(0, 2*np.pi, N, endpoint=False)
width = 2*np.pi / (N * 2)  # 条形的宽度

# 多组数据
values1 = [4, 5, 3, 6, 4, 5, 3, 6]
values2 = [3, 4, 5, 4, 6, 3, 5, 4]
values3 = [5, 3, 4, 5, 3, 6, 4, 5]

# 绘制条形
bars1 = ax.bar(theta - width, values1, width=width, 
               bottom=0, alpha=0.8, label='Q1', color='#FF6B6B')
bars2 = ax.bar(theta, values2, width=width, 
               bottom=0, alpha=0.8, label='Q2', color='#4ECDC4')
bars3 = ax.bar(theta + width, values3, width=width, 
               bottom=0, alpha=0.8, label='Q3', color='#45B7D1')

# 设置类别标签
ax.set_xticks(theta)
ax.set_xticklabels(categories, fontsize=12, fontweight='bold')

# 设置径向网格
ax.set_rgrids([1, 2, 3, 4, 5, 6], angle=0, fontsize=10)
ax.set_rlabel_position(0)  # 径向标签位置

# 设置y轴标签
ax.set_ylabel('数值', fontsize=12, labelpad=20)

# 添加图例
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0), fontsize=12)

plt.title('极坐标条形图 - 季度数据对比', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
# 创建3D散点图
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 生成随机数据
np.random.seed(42)
n_points = 200

# 创建三个类别的数据
categories = ['A', 'B', 'C']
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']

for i, (category, color) in enumerate(zip(categories, colors)):
    # 为每个类别生成不同的分布
    x = np.random.normal(i, 1, n_points)
    y = np.random.normal(i*2, 1, n_points)
    z = np.random.normal(i*3, 1, n_points)
    
    # 点的大小基于密度变化
    sizes = np.random.uniform(20, 100, n_points)
    
    # 绘制3D散点
    ax.scatter(x, y, z, 
              s=sizes,           # 点的大小
              c=color,          # 颜色
              alpha=0.6,        # 透明度
              edgecolors='black',  # 边缘颜色
              linewidth=0.5,    # 边缘线宽
              label=category,   # 标签
              depthshade=True)  # 深度阴影

# 设置标签
ax.set_xlabel('X轴 (特征1)', fontsize=12, labelpad=10)
ax.set_ylabel('Y轴 (特征2)', fontsize=12, labelpad=10)
ax.set_zlabel('Z轴 (特征3)', fontsize=12, labelpad=10)

# 添加图例
ax.legend(loc='upper right', fontsize=12)

# 设置标题
ax.set_title('3D散点图示例', fontsize=16, fontweight='bold', pad=20)

# 调整视角
ax.view_init(elev=25, azim=45)

plt.tight_layout()
plt.show()
```

```python
# 创建漏斗图数据
stages = ['访问', '注册', '试用', '购买', '复购']
conversion_rates = [100, 60, 30, 15, 8]  # 百分比
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']

# 创建漏斗图
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 8))

# 子图1: 传统漏斗图
y_pos = np.arange(len(stages))
bar_width = conversion_rates
bars = ax1.barh(y_pos, bar_width, color=colors, edgecolor='black')

# 添加数值标签
for i, (bar, rate) in enumerate(zip(bars, conversion_rates)):
    width = bar.get_width()
    ax1.text(width/2, bar.get_y() + bar.get_height()/2,
             f'{rate}%', ha='center', va='center',
             fontsize=12, fontweight='bold', color='white')

ax1.set_yticks(y_pos)
ax1.set_yticklabels(stages, fontsize=12, fontweight='bold')
ax1.invert_yaxis()  # 反转y轴
ax1.set_xlabel('转化率 (%)', fontsize=12)
ax1.set_title('转化漏斗图', fontsize=14, fontweight='bold')

# 子图2: 对称漏斗图
ax2.barh(y_pos, bar_width, color=colors, edgecolor='black')
ax2.barh(y_pos, [-b for b in bar_width], color=colors, edgecolor='black')

# 添加数值标签
for i, (bar, rate) in enumerate(zip(bars, conversion_rates)):
    width = bar.get_width()
    ax2.text(width/2, bar.get_y() + bar.get_height()/2,
             f'{rate}%', ha='center', va='center',
             fontsize=12, fontweight='bold', color='white')
    ax2.text(-width/2, bar.get_y() + bar.get_height()/2,
             f'{rate}%', ha='center', va='center',
             fontsize=12, fontweight='bold', color='white')

ax2.set_yticks(y_pos)
ax2.set_yticklabels(stages, fontsize=12, fontweight='bold')
ax2.invert_yaxis()
ax2.set_xlabel('转化率 (%)', fontsize=12)
ax2.set_title('对称漏斗图', fontsize=14, fontweight='bold')

# 添加转换率标签
conversion_labels = []
for i in range(len(conversion_rates)-1):
    rate = conversion_rates[i+1] / conversion_rates[i] * 100
    conversion_labels.append(f'{rate:.1f}%')

for i, label in enumerate(conversion_labels):
    y_pos_label = (y_pos[i] + y_pos[i+1]) / 2
    ax1.text(conversion_rates[i] + 5, y_pos_label, f'→ {label}',
            va='center', fontsize=10, fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.5))

plt.suptitle('用户转化漏斗分析', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# 创建项目数据
data = {
    'Task': ['需求分析', '系统设计', '前端开发', '后端开发', '测试', '部署', '文档编写'],
    'Start': ['2023-01-01', '2023-01-10', '2023-01-20', '2023-02-01', 
              '2023-03-01', '2023-03-15', '2023-03-20'],
    'Finish': ['2023-01-09', '2023-01-19', '2023-01-31', '2023-02-28',
               '2023-03-14', '2023-03-25', '2023-03-31'],
    'Complete': [100, 100, 80, 60, 30, 0, 0],
    'Resource': ['PM', 'Architect', 'Frontend', 'Backend', 'QA', 'DevOps', 'Tech Writer']
}

df = pd.DataFrame(data)
df['Start'] = pd.to_datetime(df['Start'])
df['Finish'] = pd.to_datetime(df['Finish'])

# 创建甘特图
fig = px.timeline(df, 
                   x_start="Start", 
                   x_end="Finish", 
                   y="Task",
                   color="Resource",  # 按资源着色
                   hover_name="Task",
                   hover_data={"Complete": ":.0f%", "Resource": True},
                   title="项目进度甘特图",
                   color_discrete_sequence=px.colors.qualitative.Set3)

# 更新布局
fig.update_layout(
    xaxis_title="时间",
    yaxis_title="任务",
    legend_title="负责人",
    width=1000,
    height=500,
    font=dict(size=12)
)

# 添加完成度文本
for i, row in df.iterrows():
    midpoint = row['Start'] + (row['Finish'] - row['Start']) / 2
    fig.add_annotation(
        x=midpoint,
        y=row['Task'],
        text=f"{row['Complete']}%",
        showarrow=False,
        font=dict(color="white" if row['Complete'] > 50 else "black"),
        yshift=0
    )

fig.show()
```
